Before training, we'll preprocess tha data by reshaping it into the shape the network expects and scaling it so that the values are in the [0,1] interval.previously our trainig images were stored in an array of the shape (60000,28,28) of type unit8 with values in range [0,255] We will transform it into float32 array of shape (60000,28*28) with the values between 0 and 1

Preparing the image data

reshape the data .... 0 to 1 interval we have to make it so that it accepts in neural....the original interval is 0 to 255

```python
train_images = train_images.reshape((60000,28*28))
train_images = train_images.astype('float32')/255
```

```python
test_images = test_images.reshape((10000,28*28))
test_images = test_images.astype('float32')/255
```

```python
from keras.utils import to_categorical
```

```python
train_labels = to_categorical(train_labels)
test_labels = to_categorical(test_labels)
```

```python
train_labels
```

#### Building nuero network

```python
from keras import models
from keras import layers
network = models.Sequential()
```

```python
network.add(layers.Dense(512,activation='relu',input_shape=(28*28,))) #dense - fully connected....512 is number of nuerons in
#layer   .... only one hidden layer
network.add(layers.Dense(10,activation='softmax')) #output layer softmax probability for each class and output the class
# with higher probability

#this is building network.
```

<img src="./march19.png">

```python
#rmsprop is used initially ...other optimisers are adam optimiser
network.compile(optimizer='rmsprop',
               loss='categorical_crossentropy',
               metrics=['accuracy'])   
```

```python
history = network.fit(train_images, train_labels, epochs=5, batch_size=128)
```

We quickly reach the accuracy 98.18% in trainid data...let us see if the model performs well in testing data too:)

```python
hist = network.evaluate(test_images,test_labels)
```

```python
train_images.shape
```

runing the same program frequesntly,,,model memorises the data.....default batch_size is 32.....we use 128 or 512

```python
print("testing accuracy",hist)
```

```python
import matplotlib.pyplot as plt
```

accuracy of training is stored in history set

```python
plt.plot(history.history['accuracy']) #history is container where we stored accuracy of train
plt.title('Model accuracy')
plt.ylabel('Accuracy')
plt.xlabel('epoch')
plt.legend(['train','test'],loc='upper left')
plt.show()
```

```python
plt.plot(history.history['accuracy'])
#history is container where we stored accuracy of train
plt.plot(hist)
plt.title('Model accuracy')
plt.ylabel('Accuracy')
plt.xlabel('epoch')
plt.legend(['train','test'],loc='upper left')
plt.show()
```

```python
history1 = network.fit(train_images, train_labels, validation_split=0.33, epochs=7, batch_size=128)
```
